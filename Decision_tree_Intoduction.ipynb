# Decision Tree 

There are various algorithms in Machine learning, so choosing the best algorithm for the given dataset and problem is the main point to remember while creating a machine learning model.

Below are the two reasons for using the Decision tree:

1. Decision Trees usually mimic human thinking ability while making a decision, so it is easy to understand.
2. The logic behind the decision tree can be easily understood because it shows a tree-like structure.

Decision Tree Terminologies
1. Root Node: Root node is from where the decision tree starts. It represents the entire dataset, which further gets divided into two or more homogeneous sets.
2. Leaf Node: Leaf nodes are the final output node, and the tree cannot be segregated further after getting a leaf node.
3. Splitting: Splitting is the process of dividing the decision node/root node into sub-nodes according to the given conditions.
4. Branch/Sub Tree: A tree formed by splitting the tree.
5. Pruning: Pruning is the process of removing the unwanted branches from the tree.
6. Parent/Child node: The root node of the tree is called the parent node, and other nodes are called the child nodes.

## Decision Tree Classifier 

Decision Tree Classifier:

1. Definition: It's a tree-like structure where internal nodes represent features, branches represent decisions, and leaf nodes represent class labels or probabilities.
2. Operation: It recursively splits the dataset based on features to create hierarchical decision rules.
3. Splitting: Nodes are split based on criteria like Gini impurity or information gain to maximize class purity at each step.
4. Predictions: It predicts the class label by traversing the tree from the root to a leaf node.
5. Advantages: Easy to understand, interpretable, and can handle non-linear relationships.
6. Disadvantages: Prone to overfitting with complex trees; sensitive to small variations in data.

```python
from sklearn.datasets import make_classification
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
from sklearn.tree import plot_tree
```

```python
# Generating synthetic data for classification
X, y = make_classification(n_samples=100, n_features=2, n_classes=2, 
                           n_clusters_per_class=1, n_redundant=0, random_state=42)
```

```python
X
```

```python
y
```

```python
# Creating a Decision Tree Classifier
clf = DecisionTreeClassifier(max_depth=3, random_state=42)
clf.fit(X, y)

# Predictions
y_pred = clf.predict(X)

# Calculating and printing accuracy score
accuracy = accuracy_score(y, y_pred)
print(f'Accuracy of decision tree classifier is : {accuracy:.2f}')
```

```python
# Plotting decision boundary
plt.figure(figsize=(8, 6))
plot_tree(clf, filled=True, rounded=True, class_names=['Class 0', 'Class 1'], feature_names=['Feature 1', 'Feature 2'])
plt.title('Decision Tree Classifier')
plt.show()
```

## Decision Tree Regressor 

Decision Tree Regressor:

1. Usage: Predicts continuous values instead of class labels.
2. Operation: Similar to classification, but instead predicts a continuous output at leaf nodes.
3. Predictions: The prediction is the average (or weighted average) of the target values of training samples falling into the leaf node.
4. Advantages: Simple to interpret, handles non-linear relationships well, and robust to outliers.
5. Disadvantages: Prone to overfitting, especially with deep trees and limited ability to capture complex patterns.

Both Decision Tree Classifier and Regressor have hyperparameters to control tree depth, node splitting criteria, and minimum samples per leaf, influencing model complexity and overfitting. Ensemble methods like Random Forests and Gradient Boosting Trees mitigate overfitting and improve performance by combining multiple decision trees. These methods are widely used in various fields due to their simplicity, interpretability, and decent performance across a range of datasets.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import r2_score
```

```python
# Generating synthetic data for regression
np.random.seed(0)
X = np.sort(5 * np.random.rand(80, 1), axis=0)
y = np.sin(X).ravel() + np.random.randn(80) * 0.1
```

```python
X
```

```python
y
```

```python
# Creating a Decision Tree Regressor
reg = DecisionTreeRegressor(max_depth=5)
reg.fit(X, y)

# Predictions
y_pred = reg.predict(X)

# Calculating and printing R-squared score
r_squared = r2_score(y, y_pred)
print(f'R-squared score of Decision Tree Regressor is : {r_squared:.2f}')
```

```python
# Plotting the results
plt.figure(figsize=(8, 6))
plt.scatter(X, y, s=20, edgecolor="black", c="darkorange", label="data")
plt.plot(X, y_pred, color="cornflowerblue", label="prediction")
plt.xlabel("data")
plt.ylabel("target")
plt.title("Decision Tree Regressor")
plt.legend()
plt.show()
```
